Verification that the main.c works as intended.

```python
import ast
import numpy as np
import matplotlib.pyplot as plt
import scipy
```

```python
with open("output_main/output_real.txt") as f:
    content_real = f.read().strip()
real = ast.literal_eval(content_real)

with open("output_main/output_imag.txt") as f:
    content_real = f.read().strip()
imag = ast.literal_eval(content_real)

with open("output_main/output_fft_real.txt") as f:
    content_real2 = f.read().strip()
fft_real = ast.literal_eval(content_real2)

with open("output_main/output_fft_imag.txt") as f:
    content_real3 = f.read().strip()
fft_imag = ast.literal_eval(content_real3)
```

```python
complex_array = np.array(real) + 1j * np.array(imag)
complex_fft_array = np.array(fft_real) + 1j * np.array(fft_imag)
```

```python
fig, axs = plt.subplots(1, 2, figsize=(16,5)) 
axs[0].plot(np.abs(complex_fft_array))
axs[0].set_title("Output from C implementation")
axs[1].plot(np.abs(scipy.fft.fft(complex_array)))
axs[1].set_title("Output from scipy fft")
plt.show()
```
